```
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```
img = cv2.imread("ganesh.jpg")
#img will be BGR which need to convert to RGB
plt.imshow(img)
```

```
#img
img2 = img.copy()
img2= cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```
#convert to grey scale
img2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
#apply Sobel for detect edges

sobelx = cv2.Sobel(img2,cv2.CV_64F,1,0,ksize=5)
#plt.imshow(sobelx)  #Doesn't look grey
plt.imshow(sobelx, cmap="gray", vmin=0, vmax=255) 
```

```
#canny another advance filter.
cImg = cv2.imread("ganesh.jpg")
cImg = cv2.cvtColor(cImg, cv2.COLOR_BGR2GRAY)
plt.imshow(cImg, cmap="gray") # vmin=0, vmax=255)
```

```
CannyImag = cv2.Canny(cImg, 100, 200)
plt.imshow(CannyImag, cmap="gray")
```
